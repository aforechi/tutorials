```lua
require 'torch'
require 'nn'
```

```lua
torch.manualSeed(123)

mlp = nn.Sequential()
mlp:add(nn.Linear(2,2))
mlp:add(nn.Sigmoid())
mlp:add(nn.Linear(2,1))
mlp:add(nn.Sigmoid())
print(mlp)
```

```lua
input = torch.Tensor({{0,0},{0,1},{1,0},{1,1}})
target = torch.Tensor({{0},{1},{1},{0}})
dataset={}
function dataset:size() 
    return 1000 
end
for i=1,dataset:size() do 
  local idx = math.random(1,4)
  local x, y = input[idx], target[idx]
  dataset[i] = {x, y}
end
```

```lua
criterion = nn.BCECriterion()

trainer = nn.StochasticGradient(mlp, criterion)
trainer.learningRate = 0.1
trainer.maxIteration = 25
trainer:train(dataset)
```

```lua
mlp:evaluate()
print(mlp:forward(input[1]))
print(mlp:forward(input[2]))
print(mlp:forward(input[3]))
print(mlp:forward(input[4]))
```

```lua
--hidden layer
print(mlp.modules[1].weight)
print(mlp.modules[1].bias)
--output layer
print(mlp.modules[3].weight)
print(mlp.modules[3].bias)
```
